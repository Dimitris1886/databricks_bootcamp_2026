# Read the data from Bronze row Data

```python
from pyspark.sql import functions as F
df = spark.table("workspace.bronze.crm_cust_info")
```

## Count rows

```python
row_count = df.count()
print(row_count)
```

## Check Business key missing 


```python
bad_key_rows = df.filter(
    F.col("cst_key").isNull() | (F.trim(F.col("cst_key")) == "")
).count()

print(bad_key_rows)
```

## Check duplicates on buisness key
##This counts "duplicate key groups" (how many keys repeat)

```python
dup_key_groups = (
    df.filter(F.col("cst_key").isNotNull() & (F.trim(F.col("cst_key")) != ""))
      .groupBy("cst_key")
      .count()
      .filter(F.col("count") > 1)
).count()
print(dup_key_groups)
```
